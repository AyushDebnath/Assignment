```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('/content/synthetic_employee_dataset.csv')
```

```python
# 1. Display the first 10 rows
print("First 10 Rows of the Dataset:\n")
print(df.head(10))
```

```python
# 2. Show complete dataset information including data types
print("Dataset Information:\n")
print(df.info())
```

```python
# 3. Number of columns and rows
print("Shape of the Dataset:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# 4a. Descriptive statistics
print("Descriptive Statistics:\n")
print(df.describe())
```

```python
# 4b. Value counts for categorical variables
print("Value Counts for Categorical Columns (including int categories):\n")
# Manually specifying numeric columns that represent categories
categorical_like_columns = ['education_level', 'department', 'certifications', 'team_size']

for col in categorical_like_columns:
    if col in df.columns:
        print(f"\nColumn: {col}")
        print(df[col].value_counts())
```

```python
# 4c. Distribution plots for numerical variables
print("Distribution Plots for Numerical Variables:\n")
numerical_columns = df.select_dtypes(include='number').columns
num_plots = len(numerical_columns)
rows = math.ceil(num_plots / 3)
cols = 3

plt.figure(figsize=(15, 5 * rows))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(rows, cols, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# 4d. Frequency analysis (value counts for all columns)
print("Frequency Analysis:\n")
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```
