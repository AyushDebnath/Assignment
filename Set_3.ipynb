```python
# Q1. Use matplotlib to create a pie chart showing the proportion of total news sold by each month.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('/content/SET_3.csv')
```

```python
month = df['month']
news_sold = df['news_sold']
```

```python
# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(news_sold, labels=month, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Total News Sold by Each Month')
plt.axis('equal')
```

```python
# Q2. Add a column named "cumulative_sales" that contains the running total of news sold up to each month.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/SET_3.csv')
```

```python
# Adding columns with running total of news sold
df['cumulative_sales'] = df['news_sold'].cumsum()
print(df)
```

```python
# Q3.  Replace the month names with their corresponding month numbers (e.g., January = 1, February = 2, etc.) in a new column.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/SET_3.csv')
```

```python
# Now to map months to numbers
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}
# Create new column with corresponding month numbers
df['month_number'] = df['month'].map(month_mapping)
print(df)
```

```python
# Q4. Identify and display the three months with the lowest news sold values.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/SET_3.csv')
```
